```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Histogram calculation/plotting in OpenCV

image = cv2.imread('assets/test.jpg')

# find pixels in range 0-255
histr = cv2.calcHist([image], [0], None, [256], [0,256])

plt.plot(histr)
plt.show()
```

```python
# Histogram calculation/plotting in Numpy
image = cv2.imread('assets/test.jpg')

# find pixels in range 0-255
histr, bins = np.histogram(image.ravel(), 256, [0,256])

plt.plot(histr)
plt.show()
```

```python
image = cv2.imread('assets/test.jpg')
color = ('b', 'g', 'r')

for i, col in enumerate(color):
    hist = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)

plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.title('BGR Histograms')
plt.grid(True)
plt.show()
```

```python
# Load the image
image = cv2.imread('assets/test.jpg')

# Create a mask (assuming a binary mask in this example)
mask = np.zeros_like(image, dtype=np.uint8)
cv2.rectangle(mask, (100, 100), (300, 300), (255, 255, 255), -1)  # rectangular mask

# Apply the mask to the image using bitwise AND operation
masked_image = cv2.bitwise_and(image, mask)

# Display the original image and the masked image
cv2.imshow('Original Image', image)
cv2.imshow('Masked Image', masked_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
