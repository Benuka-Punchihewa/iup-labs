```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('assets/dark_image.jpg')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
histr = cv2.calcHist([gray_image], [0], None, [256], [0,256])

# Equalize the histogram
equalized_image = cv2.equalizeHist(gray_image)
equalized_histr = cv2.calcHist([equalized_image], [0], None, [256], [0,256])

result = np.hstack((gray_image, equalized_image))
cv2.imshow('Result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display the original and equalized images side by side
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(histr)

plt.subplot(1, 2, 2)
plt.plot(equalized_histr)

plt.show()
```

```python
image = cv2.imread('assets/dark_image.jpg')

# Convert image from RGB to HSV
img_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Histogram equalization on the V-channel
img_hsv[:, :, 2] = cv2.equalizeHist(img_hsv[:, :, 2])

# Convert image back from HSV to RGB
equalized_image = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR)

# Display the original and equalized images
result = np.hstack((image, equalized_image))
cv2.imshow('Result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()

# plot histograms
color = ('b', 'g', 'r')
plt.figure(figsize=(12, 6))

# input image
plt.subplot(1, 2, 1)
for i, col in enumerate(color):
    hist = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)

plt.subplot(1, 2, 2)
for i, col in enumerate(color):
    hist = cv2.calcHist([equalized_image], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)

plt.show()
```
